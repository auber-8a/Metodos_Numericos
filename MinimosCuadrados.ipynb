|-----------|-----|-----|-----|-----|-----|-----|-----|-----|-----|-----|

## <span style="color:orange"> *a. Construya el polinomio por mínimos cuadrados de grado 1 y calcule el error.*     

```python
import numpy as np

valores_x = np.array([4.0, 4.2, 4.5, 4.7, 5.1, 5.5, 5.9, 6.3, 6.8, 7.1])
valores_y = np.array([102.56, 130.11, 113.18, 142.05, 167.53, 195.14, 224.87, 256.73, 299.50, 326.72])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error


coeficientes = np.polyfit(valores_x, valores_y, 1)
polinomio_grado_1 = np.poly1d(coeficientes)
predicciones_y = polinomio_grado_1(valores_x)
error_cuadratico_medio = mean_squared_error(valores_y, predicciones_y)

plt.figure(figsize=(10, 6))
plt.scatter(valores_x, valores_y, color='coral', label='Datos')
x_ajuste = np.linspace(min(valores_x), max(valores_x), 100)
plt.plot(x_ajuste, polinomio_grado_1(x_ajuste), label=f'Polinomio grado 1, Error: {error_cuadratico_medio:.2f}\nCoef: {coeficientes}')
plt.xlabel('valores_x')
plt.ylabel('valores_y')
plt.legend()
plt.title('Ajuste lineal')
plt.show()
```

**Error:** 105.88      
**Coeficientes:** [71.6102 - 191.5724]

## <span style="color:orange"> *d. Construya el polinomio por mínimos cuadrados de la forma $𝑏𝑒^{ax}$ y calcule el error.*     

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

log_valores_y = np.log(valores_y)
coeficientes = np.polyfit(valores_x, log_valores_y, 1)
a_exp, log_b = coeficientes
b_exp = np.exp(log_b)
predicciones_y_exp = b_exp * np.exp(a_exp * valores_x)
error_cuadratico_medio_exp = mean_squared_error(valores_y, predicciones_y_exp)

plt.figure(figsize=(10, 6))
plt.scatter(valores_x, valores_y, color='brown', label='Datos')
x_ajuste = np.linspace(min(valores_x), max(valores_x), 100)
y_ajuste_exp = b_exp * np.exp(a_exp * x_ajuste)
plt.plot(x_ajuste, y_ajuste_exp, label=f'Polinomio de la forma be^(ax), Error: {error_cuadratico_medio_exp:.2f}\nCoef: a={a_exp:.4f}, b={b_exp:.4f}')
plt.xlabel('valores_x')
plt.ylabel('valores_y')
plt.legend()
plt.title('Ajuste exponencial')
plt.show()
```

Grafique estos datos y encuentre la ecuación de la recta por mínimos cuadrados para estos datos.* </span>

| Puntuación ACT | Promedio de puntos | Puntuación ACT | Promedio de puntos |
|----------------|--------------------|----------------|--------------------|
| 28             | 3.84               | 29             | 3.75               |
| 25             | 3.21               | 28             | 3.65               |
| 28             | 3.23               | 27             | 3.87               |
| 27             | 3.63               | 29             | 3.75               |
| 28             | 3.75               | 21             | 1.66               |
| 33             | 3.20               | 28             | 3.12               |
| 28             | 3.41               | 26             | 2.96               |
| 29             | 3.38               | 30             | 3.10               |
| 23             | 3.53               | 24             | 2.81               |

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import mean_squared_error


puntuaciones_act = np.array([28, 25, 28, 27, 28, 33, 28, 29, 23, 27, 29, 28, 27, 29, 21, 28, 28, 26, 30, 24])
promedios_puntos = np.array([3.84, 3.21, 3.23, 3.63, 3.75, 3.20, 3.41, 3.38, 3.53, 2.03, 3.75, 3.65, 3.87, 3.75, 1.66, 3.12, 2.96, 2.92, 3.10, 2.81])


coeficientes = np.polyfit(puntuaciones_act, promedios_puntos, 1)
polinomio = np.poly1d(coeficientes)


predicciones_promedios = polinomio(puntuaciones_act)


error_cuadratico_medio = mean_squared_error(promedios_puntos, predicciones_promedios)


plt.scatter(puntuaciones_act, promedios_puntos, label='Datos', color='aqua')
plt.plot(puntuaciones_act, predicciones_promedios, label=f'Regresión lineal\n$y = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}$', color='purple')
plt.xlabel('Puntuación ACT')
plt.ylabel('Promedio de puntos')
plt.title('Regresión lineal de Puntuación ACT vs Promedio de puntos')
plt.legend()
plt.grid(True)
plt.show()


coeficientes, error_cuadratico_medio
```

## <span style="color:orange"> *4. El siguiente conjunto de datos, presentado al Subcomité Antimonopolio del Senado, muestra las características comparativas de supervivencia durante un choque de automóviles de diferentes clases. Encuentre la recta por mínimos cuadrados que aproxima estos datos (la tabla muestra el porcentaje de vehículos que participaron en un accidente en los que la lesión más grave fue fatal o seria).* </span>

| Tipo                          | Peso promedio | Porcentaje de presentación |
|-------------------------------|---------------|----------------------------|
| Regular lujoso doméstico      | 4800 lb       | 3.1%                       |
| Regular intermediario doméstico| 3700 lb       | 4.0%                       |
| Regular económico doméstico   | 3400 lb       | 5.2%                       |
| Compacto doméstico            | 2800 lb       | 6.4%                       |
| Compacto extranjero           | 1900 lb       | 9.6%                       |


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error


peso_promedio = np.array([4800, 3700, 3400, 2800, 1900])
porcentaje_presentacion = np.array([3.1, 4.0, 5.2, 6.4, 9.6])


coeficientes = np.polyfit(peso_promedio, porcentaje_presentacion, 1)
polinomio = np.poly1d(coeficientes)


predicciones_porcentaje = polinomio(peso_promedio)


error_cuadratico_medio = mean_squared_error(porcentaje_presentacion, predicciones_porcentaje)


plt.scatter(peso_promedio, porcentaje_presentacion, label='Datos', color='coral')
plt.plot(peso_promedio, predicciones_porcentaje, label=f'Regresión lineal\n$y = {coeficientes[0]:.4f}x + {coeficientes[1]:.4f}$', color='lime')
plt.xlabel('Peso promedio (lb)')
plt.ylabel('Porcentaje de presentación')
plt.title('Regresión lineal de Peso promedio vs Porcentaje de presentación')
plt.legend()
plt.grid(True)
plt.show()


coeficientes, error_cuadratico_medio
```
