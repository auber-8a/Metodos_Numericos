## <span style="color:orange"> *3. Utilice el algoritmo de eliminación gaussiana para resolver, de ser posible, los siguientes sistemas lineales, y determine si se necesitan intercambios de fila:* </span>


### <span style="color:orange"> *a. $$x_1 - x_2 + 3x_3 = 2$$ $$3x_1 - 3x_2 + x_3 = -1$$ $$x_1 + x_2 = 3$$*   

### <span style="color:orange"> *b. $$2x_1 - 1.5x_2 + 3x_3 = 1$$ $$-x_1 + 2x_3 = 3$$ $$4x_1 - 4.5x_2 + 5x_3 = 1$$*   

### <span style="color:orange"> *c. $$2x_1 = 3$$ $$x_1 + 1.5x_2 = 4.5$$ $$-3x_2 + 0.5x_3 = -6.6$$ $$2x_1 - 2x_2 + x_3 + x_4 = 0.8$$*   

### <span style="color:orange"> *d. $$x_1 + x_2 + x_4 = 2$$ $$2x_1 + x_2 - x_3 + x_4 = 1$$ $$4x_1 - x_2 - 2x_3 + 2x_4 = 0$$ $$3x_1 - x_2 - x_3 + 2x_4 = -3$$*   

```python
import numpy as np
import logging

def eliminacion_gaussiana(A: np.ndarray, b: np.ndarray) -> np.ndarray:
    """Resuelve un sistema de ecuaciones lineales mediante el método de eliminación gaussiana.

    ## Parameters

    ``A``: matriz de coeficientes del sistema de ecuaciones lineales. Debe ser de tamaño n-by-n, donde n es el número de incógnitas.
    ``b``: vector de términos independientes. Debe ser de tamaño n.

    ## Return

    ``solucion``: vector con la solución del sistema de ecuaciones lineales.

    """
    if not isinstance(A, np.ndarray):
        logging.debug("Convirtiendo A a numpy array.")
        A = np.array(A)
    if not isinstance(b, np.ndarray):
        logging.debug("Convirtiendo b a numpy array.")
        b = np.array(b)
    assert A.shape[0] == A.shape[1], "La matriz A debe ser de tamaño n-by-n."
    assert A.shape[0] == b.shape[0], "El vector b debe tener el mismo número de filas que A."
    
    n = A.shape[0]
    A = np.hstack((A, b.reshape(-1, 1)))  # Crear la matriz aumentada

    for i in range(0, n - 1):  # loop por columna

        # --- encontrar pivote
        p = None  # default, first element
        for pi in range(i, n):
            if A[pi, i] == 0:
                # must be nonzero
                continue

            if p is None:
                # first nonzero element
                p = pi
                continue

            if abs(A[pi, i]) < abs(A[p, i]):
                p = pi

        if p is None:
            # no pivot found.
            raise ValueError("No existe solución única.")

        if p != i:
            # swap rows
            logging.debug(f"Intercambiando filas {i} y {p}")
            _aux = A[i, :].copy()
            A[i, :] = A[p, :].copy()
            A[p, :] = _aux

        # --- Eliminación: loop por fila
        for j in range(i + 1, n):
            m = A[j, i] / A[i, i]
            A[j, i:] = A[j, i:] - m * A[i, i:]

        logging.info(f"\n{A}")

    if A[n - 1, n - 1] == 0:
        raise ValueError("No existe solución única.")

    print(f"\n{A}")
    # --- Sustitución hacia atrás
    solucion = np.zeros(n)
    solucion[n - 1] = A[n - 1, n] / A[n - 1, n - 1]

    for i in range(n - 2, -1, -1):
        suma = 0
        for j in range(i + 1, n):
            suma += A[i, j] * solucion[j]
        solucion[i] = (A[i, n] - suma) / A[i, i]

    return solucion
```

```python
A_a = np.array([
    [1, -1, 3],
    [3, -3, 1],
    [1, 1, 0]
], dtype=float)
b_a = np.array([2, -1, 3], dtype=float)

try:
    x_a = eliminacion_gaussiana(A_a, b_a)
    print("Solución:", x_a)
except ValueError as e:
    print("Error en (a):", e)
```

```python
A_c = np.array([
    [2, 0, 0, 0],
    [1, 1.5, 0, 0],
    [0, -3, 0.5, 0],
    [2, -2, 1, 1]
], dtype=float)
b_c = np.array([3, 4.5, -6.6, 0.8], dtype=float)

try:
    x_c = eliminacion_gaussiana(A_c, b_c)
    print("Solución:", x_c)
except ValueError as e:
    print("Error en (c):", e)
```

```python
A_d = np.array([
    [1, 1, 0, 1],
    [2, 1, -1, 1],
    [4, -1, -2, 2],
    [3, -1, -1, 2]
], dtype=float)
b_d = np.array([2, 1, 0, -3], dtype=float)


try:
    x_d = eliminacion_gaussiana(A_d, b_d)
    print("Solución:", x_d)
except ValueError as e:
    print("Error en (d):", e)
```

## <span style="color:orange"> *4.  Use el algoritmo de eliminación gaussiana y la aritmética computacional de precisión de 32 bits para resolver los siguientes sistemas lineales.* </span>



```python
def gauss(A: np.ndarray) -> np.ndarray:
    A = np.array(A)
    
    assert A.shape[0] == A.shape[1] - 1, "La matriz A debe ser de tamaño n-by-(n+1)."
    n = A.shape[0]

    for i in range(0, n - 1): 


        p = None 
        for pi in range(i, n):
            if A[pi, i] == 0:
                continue

            if p is None:
                p = pi
                continue

            if abs(A[pi, i]) < abs(A[p, i]):
                p = pi

        if p is None:
            raise ValueError("No existe solución única.")

        if p != i:
            print(f"Intercambiando filas {i} y {p}")
            _aux = A[i, :].copy()
            A[i, :] = A[p, :].copy()
            A[p, :] = _aux

        for j in range(i + 1, n):
            m = A[j, i] / A[i, i]
            A[j, i:] = A[j, i:] - m * A[i, i:]

    if A[n - 1, n - 1] == 0:
        raise ValueError("No existe solución única.")


    solucion = np.zeros(n, dtype=np.float32)
    solucion[n - 1] = A[n - 1, n] / A[n - 1, n - 1]

    for i in range(n - 2, -1, -1):
        suma = 0
        for j in range(i + 1, n):
            suma += A[i, j] * solucion[j]
        solucion[i] = (A[i, n] - suma) / A[i, i]

    return solucion
```

## <span style="color:orange"> *5. Dado el sistema lineal:* </span>

$x_1 - x_2 + \alpha x_3 = -2$  $$-x_1 + 2x_2 - \alpha x_3 = 3$$ $$\alpha x_1 + x_2 + x_3 = 2$$     
      

a. Encuentre el valor(es) de 𝛼 para los que el sistema no tiene soluciones.     
b. Encuentre el valor(es) de 𝛼 para los que el sistema tiene un número infinito de soluciones.    
c. Suponga que existe una única solución para una a determinada, encuentre la solución.    


```python
def analyze_system_with_alpha():
    from sympy import symbols, Eq, solve

    alpha = symbols('alpha')
    x1, x2, x3 = symbols('x1 x2 x3')

    equations = [
        Eq(x1 - x2 + alpha * x3, -2),
        Eq(-x1 + 2 * x2 - alpha * x3, 3),
        Eq(alpha * x1 + x2 + x3, 2)
    ]

    sol = solve(equations, (x1, x2, x3, alpha), dict=True)
    alpha_values = [s[alpha] for s in sol]

    for val in alpha_values:
        if val != alpha:
            print(f"Para α = {val}, el sistema tiene solución.")
        else:
            print("Para valores de α distintos a los encontrados, el sistema no tiene solución.")

analyze_system_with_alpha()
```

## <span style="color:orange"> *6.Suponga que en un sistema biológico existen n especies de animales y m fuentes de alimento. Si 𝑥𝑗 representa la población de las j-ésimas especies, para cada 𝑗 = 1, ⋯ , 𝑛; 𝑏𝑖; representa el suministro diario disponible del i-ésimo alimento y 𝑎𝑖𝑗 representa la cantidad del i-ésimo alimento representa un equilibrio donde existe un suministro diario de alimento para cumplir con precisión con el promedio diario de consumo de cada especie.* </span>    

a. Si        
$$
A = [a_{ij}] = \begin{bmatrix}
1 & 2 & 0 & 3 \\
1 & 0 & 2 & 2 \\
0 & 0 & 1 & 1
\end{bmatrix}
$$
     

𝐱 = ($𝑥_𝑗$) = [1000, 500, 350, 400], y 𝐛 = ($𝑏_𝑖$) = [3500, 2700, 900]. ¿Existe suficiente alimento para satisfacer el consumo promedio diario?    
b. ¿Cuál es el número máximo de animales de cada especie que se podría agregar de forma individual al sistema con el suministro de alimento que cumpla con el consumo?    
c. Si la especie 1 se extingue, ¿qué cantidad de incremento individual de las especies restantes se podría soportar?     
d. Si la especie 2 se extingue, ¿qué cantidad de incremento individual de las especies restantes se podría soportar?    


```python
def system_of_species():
    A = np.array([
        [1, 2, 0, 3],
        [1, 0, 2, 2],
        [0, 0, 1, 1]
    ])
    x = np.array([1000, 500, 350, 400])
    b = np.array([3500, 2700, 900])

    # Verificar si hay suficiente alimento
    suficiente_alimento = np.all(np.dot(A, x) <= b)
    print(f"¿Hay suficiente alimento? {'Sí' if suficiente_alimento else 'No'}")
    print(np.dot(A, x))
     
    # Número máximo de animales que se podría agregar
    incrementos_maximos = b - np.dot(A, x)
    print(f"Incrementos máximos posibles: {incrementos_maximos}")

    # Incrementos si la especie 1 se extingue
    A_ext1 = A[:, 1:]
    incrementos_ext1 = b - np.dot(A_ext1, x[1:])
    print(f"Incrementos posibles si la especie 1 se extingue: {incrementos_ext1}")

    # Incrementos si la especie 2 se extingue
    A_ext2 = np.delete(A, 1, axis=1)
    incrementos_ext2 = b - np.dot(A_ext2, np.delete(x, 1))
    print(f"Incrementos posibles si la especie 2 se extingue: {incrementos_ext2}")

system_of_species()
```

## <span style="color:orange"> *7.Repita el ejercicio 4 con el método Gauss-Jordan* </span>   

```python
def gauss_jordan(A, b):
    n = len(b)
    A = A.astype(float)
    b = b.astype(float).reshape(-1, 1)
    augmented_matrix = np.hstack((A, b))

    for i in range(n):
        augmented_matrix[i] = augmented_matrix[i] / augmented_matrix[i, i]
        for j in range(n):
            if i != j:
                augmented_matrix[j] -= augmented_matrix[i] * augmented_matrix[j, i]

    return augmented_matrix[:, -1]
```
